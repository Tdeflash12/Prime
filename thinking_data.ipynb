```python
import json
```

```python
# Load the data
def load_data(filename):
    with open(filename,'r') as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {
        'one': 1,
        'two': 2,
        'three': 3,
        'four': 4,
        'five': 5
    }

    cleaned_data = []
    unique_users = set()

    for user in data:
        # Clean ratings
        raw_rating = user.get("rating")

        if isinstance(raw_rating, str):
            raw_rating = raw_rating.lower().strip()
            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        # Handle missing age
        if user.get("age") is None:
            user["age"] = None

        # Deduplication
        name = user.get("name")

        if not name:  # skip missing names
            continue

        name = name.strip().lower()

        if name in unique_users:
            continue

        unique_users.add(user['name'])
        cleaned_data.append(user)

    return cleaned_data
```

```python
data=clean_data(data)
data
```

```python
def get_insights(data):
    # Average rating
    tot_rating = 0
    count = 0

    for user in data:
        rating = user['rating']

        if rating not in [None, '']:
            tot_rating += float(rating)
            count += 1

    if count:
        print(f'avg rating = {tot_rating/count:.2f}')
    else:
        print("No valid ratings found")

    # Poor ratings
    poor_ratings = 0

    for user in data:
        rating = user['rating']

        if rating not in [None, '']:
            if float(rating) < 3:
                poor_ratings += 1

    if count:
        print(f'% of users with poor rating = {(poor_ratings/count)*100:.2f}%')
```

```python
get_insights(data)
```

```python
def get_recommendations(data):
    recommendations = []

    for user in data:
        rating = user.get('rating')

        if rating in [None, '']:
            continue

        brand = 'Apple' if float(rating) >= 4 else 'Samsung'

        recommendations.append({
            'name': user['name'],
            'brand': brand
        })

    return recommendations
```

```python
get_recommendations(data)
```
