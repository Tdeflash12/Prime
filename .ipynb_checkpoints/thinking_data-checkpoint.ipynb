```python
import json
```

```python
# Load the data
def load_data(filename):
    with open(filename,'r') as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {
        'one': 1,
        'two': 2,
        'three': 3,
        'four': 4,
        'five': 5
    }

    cleaned_data = []
    unique_users = set()

    for user in data:
        # Clean ratings
        raw_rating = user.get("rating")

        if isinstance(raw_rating, str):
            raw_rating = raw_rating.lower().strip()
            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        # Handle missing age
        if user.get("age") is None:
            user["age"] = None

        # Deduplication
        name = user.get("name")

        if not name:  # skip missing names
            continue

        name = name.strip().lower()

        if name in unique_users:
            continue

        unique_users.add(user['name'])
        cleaned_data.append(user)

    return cleaned_data
```

```python
data=clean_data(data)
data
```

```python
#Get meaningful insights from data
def get_insights(data):

    #avg rating
    tot_rating=0
    for user in data:
        tot_rating += float(user['rating'])

    print(f'avg rating ={tot_rating/len(data)}')
```

```python
get_insights(data)
```
