```python
#filtering of data
# Filter rows where AQI (Air Quality Index) is greater than 100
df[df["aqi"] > 100]

# Filter rows where AQI is greater than 100 and temperature is above 30°C
df[(df["aqi"] > 100) & (df["temperature"] > 30)]

# Filter rows where AQI is greater than 100 and temperature is below 30°C
df[(df["aqi"] > 100) & (df["temperature"] < 30)]

# Create a new DataFrame containing only the city and AQI columns
# for locations where AQI > 100 and temperature > 30°C
aqi = df[(df["aqi"] > 100) & (df["temperature"] > 30)][["city", "aqi"]]

# Retrieve the first row of the filtered DataFrame using integer position
aqi.iloc[0]

# Retrieve the row with index label 6 from the filtered DataFrame
aqi.loc[6]

#Query -- copy
aqi_val=100
df.query("aqi>100 & temperature>30")[["city","aqi"]]
aqi_data=df.query("aqi > @aqi_val & temperature > 30")[["city", "aqi"]].copy()
```
